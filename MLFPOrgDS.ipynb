```python
# import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
import zipfile
with zipfile.ZipFile('content/creditcard.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('content/')
```

```python
data=pd.read_csv('content/creditcard.csv')
```

```python
#Fradulent/ Non-Fradulent Samples Plot
y = data["Class"]
sns.countplot(y)
target_temp = data.Class.value_counts()
print(target_temp)
```

```python
#Split the dataset into train and test set
from sklearn.model_selection import train_test_split

predictors = data.drop("Class",axis=1)
target = data["Class"]

X_train,X_test,Y_train,Y_test = train_test_split(predictors,target,test_size=0.20,random_state=0)
```

```python
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
#Random Forest
rf = RandomForestClassifier()
import time
st=time.time()
rf.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_rf = rf.predict(X_test)
et=time.time()
print("Pred Time:",et-st)
print(classification_report(Y_test,Y_pred_rf))
```

```python
#Decision Tree
dt = DecisionTreeClassifier()
import time
st=time.time()
dt.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_dt = dt.predict(X_test)
et=time.time()
print("Pred Time:",et-st)
print(classification_report(Y_test,Y_pred_dt))
```

```python
#LDA
lda = LinearDiscriminantAnalysis()
import time
st=time.time()
lda.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_lda = lda.predict(X_test)
et=time.time()
print("Pred Time:",et-st)
print(classification_report(Y_test,Y_pred_lda))
```

```python
#KNN
knn = KNeighborsClassifier()
import time
st=time.time()
knn.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_knn = knn.predict(X_test)
et=time.time()
print("Pred Time:",et-st)
print(classification_report(Y_test,Y_pred_knn))
```

```python
#Logistic Regression
lr = LogisticRegression()
import time
st=time.time()
lr.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_lr = lr.predict(X_test)
et=time.time()
print("Pred Time:",et-st)
print(classification_report(Y_test,Y_pred_lr))
```

```python
#Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
import time
st=time.time()
nb.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_nb = nb.predict(X_test)
et=time.time()
print("Pred time:",et-st)
print(classification_report(Y_test,Y_pred_nb))
```

```python
#AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
import time
st=time.time()
abc.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_abc = abc.predict(X_test)
et=time.time()
print("Pred time:",et-st)
print(classification_report(Y_test,Y_pred_abc))
```

```python
#XGBoost Classifier
import xgboost
xgb = xgboost.XGBClassifier(objective="binary:logistic")
import time
st=time.time()
xgb.fit(X_train,Y_train)
et=time.time()
print("Training Time:",et-st)
st=time.time()
Y_pred_xgb = xgb.predict(X_test)
et=time.time()
print("Pred time:",et-st)
print(classification_report(Y_test,Y_pred_xgb))
```
